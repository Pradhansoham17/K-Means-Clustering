```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('C:/Users/soham/Downloads/glass.data', names=['Id number','RI: refractive index','Na: Sodium','Mg: Magnesium',' Al: Aluminum','Si: Silicon','K: Potassium','Ca: Calcium','Ba: Barium','Fe: Iron','Type of glass'],header=None).iloc[:, 1:]
df
```

```python
np.unique(df['Type of glass'])
#only 6 types of classes are present in the dataset not 7, labels are (1,2,3,5,6,7) , label 4 missing
```

```python
df['Type of glass'].value_counts(sort=False)
```

```python
class_distribution = df['Type of glass'].value_counts(sort=False).to_dict()
class_distribution
```

```python
glass_types = list(class_distribution.keys()) 
count = list(class_distribution.values()) 
   
fig = plt.figure(figsize = (10, 5)) 
  
# creating the bar plot 
plt.bar(glass_types, count, color ='maroon',  
        width = 0.4) 
  
plt.xlabel("Type of glass") 
plt.ylabel("No. of Samples") 
plt.title("Class Distribution !!") 
plt.show() 
```

```python
import seaborn as sns 
# Plotting the KDE Plots
colors =["green", "orange",  
         "blue", "k", 'pink','cyan' ,
         "purple", 
         "red"] 
for i in np.unique(df['Type of glass']):
    sns.kdeplot(df.loc[(df['Type of glass']==i), 'RI: refractive index'], color=colors[i], shade=True, Label='Glass Type'+str(i))
    
plt.xlabel('RI: refractive index') 
plt.ylabel('Probability Density') 
```

```python
from sklearn.preprocessing import StandardScaler
data = df.iloc[:,:-1]
scaler = StandardScaler()
scaler.fit_transform(data)
```

```python
X = pd.DataFrame(scaler.fit_transform(data),index=data.index, columns=data.columns)
X
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=6, random_state=5).fit(X)
```

```python
kmeans.labels_
```

```python
X["Cluster"]=[i for i in kmeans.labels_]
X['Type of glass']=[i for i in df['Type of glass']]
X
```

```python
X.groupby(['Cluster']).count()['Type of glass']
```

```python
X.groupby(['Cluster','Type of glass']).count()
```

```python
X.groupby(['Cluster','Type of glass']).count()['RI: refractive index']
```

```python
X.groupby(['Cluster','Type of glass']).count()['RI: refractive index'][0][2]
```

```python
count_df= pd.DataFrame({'count' : X.groupby( [ 'Cluster','Type of glass'] ).size()}).reset_index()
count_df
```

```python
pivot_df = count_df.pivot(index='Cluster', columns='Type of glass', values='count')
pivot_df = pivot_df.fillna(0).astype(int)
pivot_df
```

```python
ax = pivot_df.plot(xticks=pivot_df.index)
ylab = ax.set_ylabel('Count')
```
