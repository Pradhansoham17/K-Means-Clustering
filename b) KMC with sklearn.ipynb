```python
#import libraries
import pandas as pd
import numpy as np
import random as rd
import matplotlib.pyplot as plt
df = pd.read_csv('C:/Users/soham/Downloads/clustering.csv')
df
```

```python
from sklearn.cluster import KMeans
import numpy as np
X = df.loc[:,["LoanAmount","ApplicantIncome"]]
kmeans = KMeans(n_clusters=3, random_state=5).fit(X)
```

```python
kmeans.labels_
```

```python
X["Cluster"]=kmeans.labels_
```

```python
X
```

```python
kmeans.predict([[800,7800],[120,1500]])
```

```python
#final cluster centers after algorithm is completed
kmeans.cluster_centers_
```

```python
#Sum of squared distances of samples to their closest final cluster center
kmeans.inertia_ 
```

```python
Centroids = pd.DataFrame(kmeans.cluster_centers_, columns = ["LoanAmount","ApplicantIncome"])  
Centroids
```

```python
for i, (cx, cy) in enumerate(kmeans.cluster_centers_):
    print(i,cx,cy)
```

```python
color=['blue','green','cyan']
for k in range(3): #because n_clusters=3
    data=X[X["Cluster"]==k]
    plt.scatter(data["ApplicantIncome"],data["LoanAmount"],c=color[k], edgecolor='k')
plt.scatter(Centroids["ApplicantIncome"],Centroids["LoanAmount"],c='red')
plt.xlabel("Applicant's Income")
plt.ylabel('Loan Amount (In Thousands)')
plt.show()
```
